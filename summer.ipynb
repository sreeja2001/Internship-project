```python
## which sport is having most number of gold medals
```

```python
data=[]
for i in df['Sport'].unique():
  data.append([i,len(df[df['Sport']==i])])
pd.DataFrame(data,columns = ['Gold','Sport']).sort_values(by='Gold', ascending = False).plot(x = 'Gold', y = 'Sport', kind = 'bar', figsize = (10,5))
```

```python
# based on medals
```

```python

data=[]
for i in df['Sport'].unique():
  data.append([i,len(df[df['Sport']==i])])
pd.DataFrame(data,columns = ['Medal','Sport']).sort_values(by='Medal', ascending = False).plot(x = 'Medal', y = 'Sport', kind = 'bar', figsize = (10,5))
```

```python
df = pd.DataFrame(data, columns=['Sport', 'Medal'])
df['Sport'].loc[df['Medal'] == df['Medal'].max()]
```
